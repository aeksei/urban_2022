```python
import os

from google.colab import drive
import geopandas as gpd
import pandas as pd
import osmnx as ox
```

```python
ROOT_DIR = "/content/drive"
BASE_DIR = os.path.join(ROOT_DIR, "MyDrive", "PythonUrban", "geojson_src")
TILES = "CartoDB positron"

drive.mount(ROOT_DIR)
```

```python
filename = os.path.join(BASE_DIR, "green_zones.geojson")
green_zone_gdf = gpd.read_file(filename)
green_zone_gdf.head()
```

```python
green_zone_gdf.info()
```

```python
green_zone_gdf.crs
```

```python
green_zone_gdf.explore(color="green", tiles=TILES)
```

```python
df = pd.read_html("https://wiki.openstreetmap.org/wiki/RU:%D0%A1%D0%B0%D0%BD%D0%BA%D1%82-%D0%9F%D0%B5%D1%82%D0%B5%D1%80%D0%B1%D1%83%D1%80%D0%B3/%D0%A0%D0%B0%D0%B9%D0%BE%D0%BD%D1%8B")[0]
df
```

```python
districts_series = "Санкт-Петербург " + df["Район"] + " район"
districts_series.to_list()
```

```python
districts = ox.geocode_to_gdf(districts_series.to_list())
districts
```

```python
def get_green_zone_intersection(district_geometry):
    return green_zone_gdf.geometry.intersection(district_geometry).unary_union

districts["green_zone_intersection"] = districts.geometry.apply(get_green_zone_intersection)
districts.head()
```

```python
districts["coverage"] = districts["green_zone_intersection"].to_crs(3857).area / districts.geometry.to_crs(3857).area
districts.head()
```

```python
districts.sort_values("coverage", ascending=False, inplace=True)
districts
```

```python
districts["percent_coverage"] = districts["coverage"].apply(lambda item: f"{item:.2%}")
districts.head()
```

```python
districts["position"] = districts["coverage"].rank(ascending=False).astype(int)
districts.head()
```

```python
style_kwds = {
    "color": "black"
}

m = districts.explore("coverage", tiles=TILES, cmap="Greens", style_kwds=style_kwds)
districts["green_zone_intersection"].explore(m=m, color="green")
```
